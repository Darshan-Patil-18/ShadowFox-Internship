# **Matplotlib** Visualization Library Documentation

## Library Overview

**Matplotlib** is one of the most widely used Python libraries for data visualization.  
It is considered the foundation of plotting in Python and provides full control over graph styling and customization.

### Key Features
- Supports many types of plots (line, bar, scatter, histogram, pie, etc.)
- Highly customizable with labels, colors, styles, and themes
- Works well with NumPy and Pandas datasets
- Best suited for static and publication-quality charts

### Typical Use Cases
- Exploratory Data Analysis (EDA)
- Academic and research projects
- Data reporting and trend visualization
- Building custom visualizations from scratch


# **MATPLOTLIB PLOTS**

```python
#import matplotlib library
import matplotlib.pyplot as plt
```

## 1. Line Plot

### Description
A **line plot** displays data points connected by straight lines.
### Use Case
Used to show trends over time such as temperature change, sales growth, or performance tracking.


```python
x = [1,2,3]
y= [1,5,5]

plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

## 2.Bar plot


### Description
A **bar chart** represents categorical data using rectangular bars.

### Use Case
Used to compare values across categories such as product sales, student marks, or survey results.

```python
category = ['Male','Female','Others']
Numbers= [122332,110001,122403]
colors = ['Blue','pink','yellow']

plt.bar(category,Numbers,color=colors)
plt.title("Bar plot")
plt.xlabel('Genders')
plt.ylabel("Count")
plt.show()
```

## 3. Scatter Plot

### Description
A **scatter plot** represents data points as individual dots on a 2D graph.

### Use Case
Used to observe relationships or correlations between two numerical variables.

```python
x= [1,4,5,4,7,5,5,6,5,4,5,4,8,9]
y =[4,5,8,8,5,6,8,5,6,7,5,6,4,5]

plt.scatter(x,y,color='red')
plt.xlabel('X axis')
plt.ylabel("Y axis")
plt.title("Scatter Plot")
plt.show()
```

## 4. Histogram

### Description
A **histogram** shows the frequency distribution of numerical data.

### Use Case
Used to understand how values are spread, such as salary distribution or exam score frequency.


```python
x = [3,3,3,2,3,2,3,4,5,5,6,4,3,6,5]

plt.hist(x,bins=5,edgecolor='red')
plt.title("Histogram")
plt.xlabel("Bins distance")
plt.ylabel('Values')
plt.show()
```

## 5. Pie Chart

### Description
A **pie chart** represents data as slices of a circle showing proportions.

### Use Case
Used to visualize percentage contribution such as market share or language usage.


```python
labels = ['A','B','C','D']
sizes = [20,40,45,80]

plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title('Pie Plot')
plt.show()
```

## 6. Box Plot

### Description
A **box plot** summarizes data distribution and highlights outliers.

### Use Case
Used in statistical analysis to detect spread and unusual values in datasets.


```python
data = [34,50,23,11]

plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

## 7. Horizontal Bar Chart

### Description
A **horizontal bar chart** displays categories along the vertical axis and values along the horizontal axis.

### Use Case
Used when category names are long or when you want easier readability compared to vertical bar charts.


```python
category = ['A','B','C','D']
data = [10,20,30,40]
plt.barh(category,data,color='red')
plt.title('Horizontal BAR')
plt.show()
```

# **Seaborn** Visualization Library Documentation

## 1. Library Overview

**Seaborn** is a high-level Python visualization library built on top of Matplotlib.  
It provides a simpler interface and produces more attractive statistical graphics.

### Key Features
- Works directly with Pandas DataFrames
- Provides beautiful default themes and color palettes
- Best for statistical and machine learning visualizations
- Supports advanced plots like heatmaps, pairplots, and boxplots

### Typical Use Cases
- Exploratory Data Analysis (EDA)
- Statistical data visualization
- Correlation analysis
- Machine learning dataset visualization


# **SEABORN PLOTS**

```python
%pip install seaborn
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")

tips
```

## 1. Count Plot

### Description
A **count plot** is used to display the frequency (count) of each category in a dataset.

### Use Case
Used when you want to analyze how many times each category appears, such as:
- Number of students in each department
- Count of products in each category
- Frequency of classes in a classification dataset


```python
sns.countplot(x='smoker',data=tips)
plt.title('COUNT plot')
plt.show()
```

## 2. Histogram Plot (Histplot)

### Description
A **histplot** in Seaborn is used to visualize the distribution of a numerical dataset by grouping values into bins.

### Use Case
Used when you want to understand:
- Frequency of values in a dataset
- Data distribution (normal, skewed, etc.)
- Patterns in marks, salaries, ages, etc.


```python
sns.histplot(x='total_bill',bins=20,data=tips,kde=True)
plt.title('Hist PLOT')
plt.show()
```

## 3. Pair Plot

### Description
A **pair plot** in Seaborn is used to visualize relationships between multiple numerical variables in a dataset.

It creates scatter plots for each pair of features and histograms (or KDE plots) along the diagonal.

### Use Case
Pair plots are very useful in:
- Exploratory Data Analysis (EDA)
- Machine learning dataset understanding
- Finding correlations and clusters between variables
- Visualizing multi-feature datasets like Iris


```python
sns.pairplot(tips)
plt.show()
```

## 4. Line Plot

### Description
A **line plot** displays data points connected by straight lines.  
It is commonly used to visualize trends or changes over a continuous variable.

### Use Case
Used to show trends over time such as:
- Stock price movement
- Temperature changes
- Growth analysis


```python
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line Plot')
plt.show()
```

## 5. Box Plot

### Description
A **box plot** is a statistical plot that shows:
- Median
- Quartiles
- Spread of data
- Outliers

### Use Case
Used to compare distributions and detect outliers in datasets, such as:
- Salary distribution
- Student marks analysis
- Product price variation


```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box plot')
plt.show()
```

## 6.Violin Plot

### Description
A **violin plot** is similar to a box plot, but it also shows the density of the data distribution.

It combines:
- Box plot summary
- Probability distribution shape

### Use Case
Used when you want deeper insight into the distribution of data, such as:
- Comparing multiple groups
- Visualizing data spread and concentration


```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('Violin Plot')
plt.show()
```

# Comparison: Matplotlib vs Seaborn

| Feature | Matplotlib | Seaborn |
|--------|------------|---------|
| Ease of Use | Requires more code and manual setup | Simple syntax, beginner-friendly |
| Customization | Very high (full control over plots) | Moderate (built on Matplotlib) |
| Default Appearance | Basic unless styled manually | Attractive and modern by default |
| Statistical Plots | Limited (basic charts only) | Excellent (heatmaps, violin, pairplots) |
| DataFrame Support | Works, but needs extra effort | Direct support with Pandas DataFrames |
| Interactivity | Mostly static plots | Mostly static plots |
| Performance on Large Data | Slightly faster | Slightly slower due to extra styling |
| Best Use Case | Full customization and detailed plotting | Quick EDA and statistical visualization |

---

## Final Conclusion

- **Matplotlib** is best for complete control and custom visualizations.  
- **Seaborn** is best for quick, beautiful, and statistical plots with less code.
